Data Cleaning

```python
# Filtering object columns
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
#Replacing columns with 'yes' and'no'
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.gender.unique()
```

```python
#One hot encoding for categorical columns
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
# Scaling the data Monthlycharges, Totalcharges and tenure
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

Now the DataFrame is ready for the Model

Splitting into Train and Test dataset

```python
X = df2.drop('Churn',axis='columns')
y = df2['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train[:10]
```

Building Artificial Neural Network (ANN) Model

```python
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
#changing the 2D array
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)

y_pred[:10]
```

```python
y_test[:10]
```

Making a Report of the Model

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

The Overall Accuracy we get is 79 percent

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
import keras_tuner as kt
```

```python
def build_model(hp):
    model = Sequential()

    counter = 0
    for i in range(hp.Int('num_layers', min_value = 1, max_value=10)):
        if counter == 0:
            model.add(
                Dense(
                    hp.Int('units' + str(i),min_value = 8, max_value = 128, step = 8),
                    activation = hp.Choice('activation' + str(i), values = ['relu','tanh','sigmoid']),
                    input_dim=8
                )
            )
        else:
            model.add(
            Dense(
                hp.Int('units' + str(i),min_value = 8, max_value = 128, step = 8),
                activation = hp.Choice('activation' + str(i), values = ['relu','tanh','sigmoid']),

                )
            )
        counter+=1

    model.add(Dense(1,activation = 'sigmoid'))

    model.compile(optimizer=hp.Choice('optimizer',values=['rmsprop','adam','sgd','nadam','adadelta']),
    loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model


```

```python
tuner = kt.RandomSearch(build_model,
                        objective='val_accuracy',
                        max_trials = 3,
                        directory='mydir',
                        project_name='final')
```

```python
tuner.search(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
